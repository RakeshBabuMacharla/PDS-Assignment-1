```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv("/content/StudentsPerformance.csv")
print(df.head())  # Display first 5 rows
```

#Performing 5 Data Visualizations

#1. Histogram of Student Scores

```python
# Create the 'figures' directory if it doesn't exist
os.makedirs("figures", exist_ok=True)

plt.figure(figsize=(8,4))
sns.histplot(df["math score"], bins=10, kde=True, color="blue")
plt.xlabel("Math Score")
plt.title("Distribution of Math Scores")
plt.savefig("figures/hist_math.png")
plt.show()
```

**Purpose:**

Shows the distribution of student scores in math.
Identifies skewness, normality, and outliers.

**Analysis Made Easier:**

Helps educators understand the overall performance of students.
Identifies if the scores are normally distributed or skewed.
Helps pinpoint outliers (exceptionally high or low scores).


#2. Scatter Plot (Math vs. Writing Score)

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["math score"], y=df["writing score"])
plt.xlabel("Math Score")
plt.ylabel("Writing Score")
plt.title("Math vs Writing Score")
plt.savefig("figures/scatter_scores.png")
plt.show()
```

**Purpose:**

Identifies correlation between two subjects.
Reveals if performance in one subject predicts performance in another.

**Analysis Made Easier:**

Helps teachers identify patterns, such as whether students who excel in math also perform well in writing.
Highlights students needing subject-specific interventions if there is a gap between scores.
Useful in evaluating multi-subject competency of students.

#3. Box Plot of Scores by **Gender**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["gender"], y=df["math score"])
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.title("Math Scores by Gender")
plt.savefig("figures/boxplot_gender.png")
plt.show()
```

**Purpose:**

Compares score distribution between male and female students.
Identifies differences in performance across genders.

**Analysis Made Easier:**

Helps educators detect gender-based disparities in subject performance.
Shows the median, quartiles, and outliers, providing a detailed statistical comparison between male and female students.
Supports education policy decisions to promote gender equality in learning.

#4. Bar Chart of Average Scores by Subject

```python
avg_scores = df[["math score", "reading score", "writing score"]].mean()

plt.figure(figsize=(6,4))
sns.barplot(x=avg_scores.index, y=avg_scores.values)
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.title("Average Scores by Subject")
plt.savefig("figures/bar_avg_scores.png")
plt.show()
```

**Purpose:**

Compares the average scores of students across different subjects.
Highlights which subjects have better or worse performance overall.

**Analysis Made Easier:**

Helps identify subjects where students excel or struggle collectively.
Assists schools in allocating resources.
Enables curriculum adjustments to improve weaker subject areas.

#5. Heatmap of Correlation Matrix

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.savefig("figures/heatmap_corr.png")
plt.show()
```

**Purpose:**

Displays correlations between all numerical variables in the dataset.
Helps identify relationships between study hours, attendance, and grades.

**Analysis Made Easier:**

Shows which factors have the strongest impact on student performance.
Helps teachers and parents focus on key predictors of success.
Identifies unexpected relationships, such as whether reading scores impact math scores.
